```python
import pandas as pd
import seaborn as sns

import numpy as np
import matplotlib.pyplot as plt

file_path = "final_census_dataset(1).csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info()

# Display the first few rows
df.head()
```

```python
df_sampled = df.sample(n=1000, random_state=42)  # Set random_state for reproducibility

# Display the first few rows of the sampled dataset
df_sampled.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Checking for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Display unique values for categorical columns in a readable format
categorical_columns = ['SEX', 'RAC1P', 'MAR', 'COW', 'POBP']
print("\nUnique Values in Categorical Columns:\n")
for col in categorical_columns:
    print(f"{col}: {sorted(df[col].unique())}")
```

```python
# Simple and Clear Heatmap for Descriptive Statistics
plt.figure(figsize=(10, 6))

# Create heatmap
sns.heatmap(df.describe().T, 
            annot=True, 
            fmt=".2f", 
            cmap="coolwarm", 
            linewidths=0.5, 
            linecolor="black")

# Add title and adjust layout
plt.title("Descriptive Statistics Heatmap", fontsize=14, fontweight="bold", pad=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12, rotation=0)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
selected_features = ["AGEP", "SCHL", "WKHP", "OCCP", "Income_Above_50K"]
sns.pairplot(df[selected_features], hue="Income_Above_50K", diag_kind="kde")
plt.show()
```
